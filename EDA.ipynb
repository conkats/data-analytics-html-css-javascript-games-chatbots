##A real example

```python
import pandas as pd

import seaborn as sns#seaborn is a package that is used for visualisation applications
sns.set_theme()
sns.set(rc={'figure.figsize':(11,7)})
from matplotlib import pyplot as plt # matplotlib is another package visualisation
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/boolean-data-analytics/data/main/Superstore%20Data%20S.csv')
df.head()#categorical data, product details
```

```python
#book: https://wesmckinney.com/book/numpy-basics.html
```

##Data Analysis

```python
df['Country'].unique()#check how many countries there are
```

```python
df.info()
```

```python
us = df[df['Country'] == 'United States']
```

```python
us['Country'].unique()
```

```python
us.head(3)#row index 1 shows -ve profit, loss?
```

```python
us_cat_profit = us.groupby('Category', as_index=False)['Profit'].sum().sort_values('Profit',ascending=False)#groupy Category by Profit sum and sort by profit in ascending order i.e highest Profit at the top
us_cat_profit
plt.title("Profit by Category")
sns.barplot(x='Category', y ='Profit', data=us_cat_profit, color='grey')
plt.show()
```

```python
us.groupby('Category', as_index=False)['Profit'].sum()
```

```python
us['Profit'].mean()
```

```python
us.groupby('Category', as_index=False)['Profit', 'Sales'].sum()
```

```python
us['Profit'].sum()
```

```python
us.groupby("Region", as_index=False)['Profit'].sum()
```

```python
us_region_profit = us.groupby('Region', as_index=False)['Profit'].sum()
sns.set_color_codes(palette="deep")
sns.barplot(x='Region', y ='Profit', data=us_region_profit)
plt.show()
```

```python
us_cat_region_profit = us.groupby(['Category','Region'], as_index=False)['Profit'].sum()
sns.barplot(x='Region', y ='Profit', hue='Category', data=us_cat_region_profit)
plt.show()
#furniture losing us money in the central region
#low profit margins in the south
#the company needs to invest more in technology
```

An alternative to the bar chart can be a pivot table:

```python
pd.pivot_table(data=us_cat_region_profit,
              values='Profit',
               index='Category',
               columns = 'Region',
               aggfunc = ['sum','min']
               )
```

```python
us_cntr_furn = us[(us['Category']=='Furniture') & (us['Region']=='Central')]
#us['Category']=='Furniture'
#us['Region']=='Central')]
#AND & in pandas
#OR |

us_cntr_furn
```

```python
us_cntr_furn.shape#method that allows to know rows and columns
```

```python
sns.histplot(data=us_cntr_furn, x= 'Profit', bins=35, kde=True)#to show a distribution of single variable, x=variable, bins = how many bucket we want to break the x axis, kde = smoothness of kernel estimate density
plt.show()
```

```python
#peak of distribution is to the left of 0 so a concern
#so problematic for profit
```

```python
us_cntr_furn['Profit'].sum()
```

```python
#losing -2871 dollars
```

```python
us_cntr_furn.groupby(['Sub-Category'],as_index=False)['Profit'].sum()
```

```python
sns.boxplot(data=us_cntr_furn, x='Profit', y = 'Sub-Category')
plt.show()
#when left of zero then there is a problem
```

```python
us_cntr_furn.groupby(['Sub-Category'],as_index=False)['Discount'].mean()
```

```python
us_profit_dscnt = us_cntr_furn.groupby(['Sub-Category'], as_index=False).agg(
    profit_sum = ('Profit','sum'),
    discount_avg=('Discount','mean')
)
us_profit_dscnt
```

```python
sns.scatterplot(data=us_profit_dscnt, x='profit_sum', y='discount_avg',hue = 'Sub-Category',s = 100)
plt.show()
```

```python
#profits fall as discounts increases
```
